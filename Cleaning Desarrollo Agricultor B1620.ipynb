```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df_agri= pd.read_csv("Agricultura_y_Desarrollo_Rural_-_Evaluaci_n_Agr_cola_Departamento_de_Bol_var._20240930.csv")
```

```python
def amount(): 
    return df_agri.shape
```

```python
def info():
    return df_agri.info()

info()
```

```python
# rename columns 6, 7, 8, 9 to make them easier to read. 
df_agri.rename(columns={
    'Área Sembrada\n(ha)':'Area_sembrada_ha',
    'Área Cosechada\n(ha)':'area_cosechada_ha',
    'Producción\n(t)':'produccion_t',
    'Rendimiento\n(t/ha)':'redimiento_tha'},inplace=True)

print(df_agri.columns)
```

```python
df_agri.rename(columns={
    'area_consechad_ha': 'area_cosechada_ha',},inplace=True)

print(df_agri.columns)
```

```python
# we standardise the name of the columns
df_agri.columns=df_agri.columns.str.lower()

print (info())
```

```python
# let's take a look at the data
def encabezado():
    return df_agri.head(40)


encabezado()
```

```python
# now we will see the null data, according to what we have seen, there are only 9 apparently. 
filas_nullas= df_agri[df_agri.isnull().any(axis=1)]

print(filas_nullas)
```

```python
# Given that all the null data have values in the area sown, I will opt to replace those data with values of 0 
# taking into account that the null data value is in 9 columns and I do not consider it a relevant value. 

# filling in null values 
df_agri.fillna(0, inplace=True)

info()
```

```python
# We change the data type of columns 6,7,8,9 to float type.
# Removing commas 
df_agri['area_sembrada_ha'] = df_agri['area_sembrada_ha'].str.replace(',', '')
df_agri['area_cosechada_ha'] = df_agri['area_cosechada_ha'].str.replace(',', '')
df_agri['produccion_t'] = df_agri['produccion_t'].str.replace(',', '')


print(df_agri.dtypes)

```

```python
df_agri['area_sembrada_ha'] = df_agri['area_sembrada_ha'].str.replace(',', '').astype(float)
df_agri['area_cosechada_ha'] = df_agri['area_cosechada_ha'].str.replace(',', '').astype(float)
df_agri['produccion_t'] = df_agri['produccion_t'].str.replace(',', '').astype(float)


print(df_agri.dtypes)
```

```python
# let's do a numerical analysis to see how the quantitative data stands. 
def ana_numerico():
    return df_agri.describe()

ana_numerico()
```

```python
#Graphing numerical data 
def ana_grafico_numerico():
    col_num= ['periodo','area_sembrada_ha','area_cosechada_ha','produccion_t','redimiento_tha']
    fig, ax=plt.subplots(nrows=5,ncols=1,figsize=(8,30))
    fig.subplots_adjust(hspace=0.5)


    for i,col in enumerate(col_num):
        sns.boxplot(x=col, data=df_agri, ax=ax[i],showfliers=False)
        ax[i].set_title(col)
        print (f'columna {col}:{df_agri[col].nunique()} valores unicos')
```

```python
#called ana_graphic_number function:
ana_grafico_numerico()
```

```python
# now with all the columns in the appropriate data type 
# proceed to check the categorical variables 
def ana_categorico():
    colm_cat= [
        'departamento',
        'codmunicipio',
        'municipio',
        'cultivo'
    ]
    for col in colm_cat:
        valor= df_agri[col].nunique()
        print('Columnna', col,':',valor,'valor unico')
```

```python
ana_categorico()
```

```python
# now let's look at the unique values of these categorical columns. 
# now to better see the duplicate values in the column municipality and crop, I will make a list of them. 
def ana_grafico_categorico():
    colm_cat= [
        'departamento',
        'codmunicipio',
        'municipio',
        'cultivo'
    ]
    fig, ax = plt.subplots(nrows=4, ncols=1,figsize = (10,30))
    fig.subplots_adjust(hspace=1)

    for i,col in enumerate(colm_cat):
        sns.countplot(x=col, data=df_agri, ax=ax[i])
        ax[i].set_title(col)
        ax[i].set_xticklabels(ax[i].get_xticklabels(), rotation=90)
    # alphabetical order
    list_alfa =df_agri[['municipio','cultivo']].drop_duplicates().sort_values(by=['municipio','cultivo'])
    print(list_alfa.reset_index(drop=True))
  
```

```python
# let's change the tildes to unify the values in the crop column. 
df_agri['cultivo'] = df_agri['cultivo'].str.replace('á', 'a',regex=False)
df_agri['cultivo'] = df_agri['cultivo'].str.replace('é', 'e',regex=False)
df_agri['cultivo'] = df_agri['cultivo'].str.replace('í', 'i',regex=False)
df_agri['cultivo'] = df_agri['cultivo'].str.replace('ó', 'o',regex=False)
df_agri['cultivo'] = df_agri['cultivo'].str.replace('ú', 'u',regex=False)

# check 
list_alfa =df_agri[['cultivo']].drop_duplicates().sort_values(by=['cultivo'])
print(list_alfa.to_string(index=False))
```

```python
# ready, now as we see that some crops have the same names at the beginning we will search for these names 
# we filter by values that have a space 
filtro_espacio= df_agri[df_agri['cultivo'].str.contains(' ')]
conteo_cultivos = filtro_espacio['cultivo'].value_counts().sort_index() #  count containing the filter

print(conteo_cultivos.drop_duplicates().to_string())
print(filtro_espacio[['cultivo']].drop_duplicates().sort_values(by=['cultivo']).to_string(index=False))
```

```python
# Given the results and an investigation, it seems that the only one that is the same is sugar cane. 
# crops such as cucumber, cassava and tobacco are just a few examples where the crop variety is differentiated and left as different values. 
# so it will be unified 
df_agri['cultivo'] = df_agri['cultivo'].str.replace('caña azucarera', 'caña',regex=False)
df_agri['cultivo'] = df_agri['cultivo'].str.replace('caña panelera', 'caña',regex=False)


# check
cultivo_counts = df_agri['cultivo'].value_counts().reset_index()
cultivo_counts.columns = ['cultivo', 'count']  # Rename columns for clarity

# sort by cultivo
cultivo_counts = cultivo_counts.sort_values(by='cultivo')

print(cultivo_counts.to_string(index=False))
```

```python
# observing some outliers or very scattered values in the graphs of the numerical values
# in the rendimiento_tha column with values above 100 %.
# we will look at these values individually 
filtro_rend= df[df['rendimiento_tha']>100]
print(filtro_rend)
```

```python
# The data are recognised as miscalculated data, therefore the appropriate value for the miscalculated values is recalculated. 
#row traversal in the dataframe
for i, row in df.iterrows():
    #the condition is set to be >100
    if row['rendimiento_tha']>100: 
        #calculation of the appropriate value in the row
        df.at[i,'rendimiento_tha']= row['produccion_t']/ row['area_cosechada_ha'] 
#change check 
print(df[df['rendimiento_tha']>100][['produccion_t','area_cosechada_ha','rendimiento_tha']])




```

```python
"""Given the above, due to the inconsistency of the numbers, we proceed to eliminate all rows with a value greater than 100 in the
greater than 100 in the rendimiento_tha column """

df= df[df['rendimiento_tha']<=100]
```

```python
# we plot to visually see the arrangement of the data in the rendimiento_tha column 

# creation of the canvas 
plt.figure(figsize=(7,6))
sns.boxplot(x=df['rendimiento_tha'],showfliers=True)
plt.title('Dispersión Rendimiento_tha')
plt.xlabel('rendimiento de la producción')
plt.show()
```
