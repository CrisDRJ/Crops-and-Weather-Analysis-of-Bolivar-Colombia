```python
# Calculate the unique values of each column

# Define the function unique_column_value

def unique_column_value(df_prec):
    unique_v = {}  # Create a dictionary to store the data
    for column in df_prec.columns:
        unique_count = df_prec[column].nunique()  # Count the unique values
        unique_v[column] = unique_count  # Store the value in the dictionary

    return unique_v

# Get unique values

valores_unicos = unique_column_value(df_prec)

# Display the unique values by column

for col, vals in valores_unicos.items():
    print(f'Unique value in {col}: {vals}')
```

```python
# Proceed to remove unnecessary columns for this project
""" 
'NombreEstacion' = with the station code, we have sufficient information
'Latitud' = since the main data is grouped by department, the department column is enough for establishing the relationship
'Longitud' = since the main data is grouped by department, the department column is enough for establishing the relationship
'Altitud' = since the main data is grouped by department, the department column is enough for establishing the relationship
'AreaOperativa' = since the main data is grouped by department, the department column is enough for establishing the relationship
'FechaSuspension' = all values are null and do not represent valuable information in this project
'Calificador' = It is not relevant to the project objectives
'DescripcionSerie' = has a unique value and the frequency column provides the corresponding information
'NivelAprobacion' = It is not relevant to the project objectives
"""

# Create a list with the names of the columns to be deleted
column_delete = ['NombreEstacion', 'Latitud', 'Longitud', 'Altitud', 'AreaOperativa', 'FechaSuspension', 'Calificador', 'NivelAprobacion', 'DescripcionSerie', 'FechaInstalacion']

def delete_column_prec(df_prec):
    # Iterate over each column in the column_delete list
    for col in column_delete:
        # Check if the column is in df_prec
        if col in df_prec.columns:
            # Delete the column from the DataFrame
            df_prec = df_prec.drop(columns=[col])
            # Print deleted column
            print(f'Column deleted: {col}')
    return df_prec
            
```

```python
# Check unique values of categorical variables
# Specifically, the columns 'Categoria' and 'Municipio'
columna_Categoria = 'Categoria'

# Define a function to view each unique value
def unique_catg(df_precipitacion, columna):
    # Verify the column in the data frame
    if columna in df_precipitacion.columns:
        # Get the unique values
        unicos = df_precipitacion[columna].unique()
        # Count the number of unique values
        cantidad_unicos = len(unicos)
        # Print each unique value
        for i in unicos:
            print(f'*{i}')
        print(f'The number of unique values in {columna} is: {cantidad_unicos}')
```

```python
# We check the unique values in the column 'Categoria'
unique_catg(df_precipitacion,columna_Categoria)
```

```python
# We check the unique values in the column'Municipio'
Columna_Municipio= 'Municipio'
unique_catg(df_precipitacion,Columna_Municipio)
```

```python
# taking into account that in the main table each municipality's data is without a tilde, the change is made
df_precipitacion['Municipio']= df_precipitacion['Municipio'].str.replace('í','i')
df_precipitacion['Municipio']= df_precipitacion['Municipio'].str.replace('á','a')
df_precipitacion['Municipio']= df_precipitacion['Municipio'].str.replace('é','e')
df_precipitacion['Municipio']= df_precipitacion['Municipio'].str.replace('ó','o')
df_precipitacion['Municipio']= df_precipitacion['Municipio'].str.replace('ú','u')

```

```python
#we verify the changes with the unique_catg function

unique_catg(df_precipitacion,Columna_Municipio)
```

```python
# Let's view precipitation data by year graphically

# Change data type of the 'Fecha' column
df_precipitacion['Fecha'] = pd.to_datetime(df_precipitacion['Fecha'])

# Create the Año column
df_precipitacion['Año'] = df_precipitacion['Fecha'].dt.year

# Group by year and calculate the average precipitation
df_group_A_P = df_precipitacion.groupby('Año', as_index=False)['Valor'].mean()

# Create line plot of Año vs. Precipitation

# Set up the canvas
plt.figure(figsize=(10,6))

# Set up the plot
sns.lineplot(x='Año', y='Valor', data=df_group_A_P)

# Design the plot
plt.title('Average Precipitation by Year', fontsize=18)
plt.xlabel('Year')
plt.ylabel('Precipitation')

# Display
plt.grid()
plt.show()
```
